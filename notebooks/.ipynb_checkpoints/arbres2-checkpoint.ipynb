```python
import csv
with open('/home/dataplus-01/Documents/Modules_Socle/Python/Projet_arbre/data/trees.csv', newline="") as f:
    reader=csv.reader(f)
    data=list(reader)
```

```python
# Pour commencer, identifiez tous les genres botaniques représentés à Grenoble.
# Combien y en a-t-il ?
data_GENRE_BOTA=[]
ind=data[0].index('GENRE_BOTA')
for e in data[1:]:
    if e[ind]!="":
        data_GENRE_BOTA.append(e[ind])
set_GENRE_BOTA=set(data_GENRE_BOTA)
print(len(set_GENRE_BOTA)) #sans compter le genre vide "" 
```

```python
# Essayez plusieurs manières de résoudre ce problème.
```

```python
# Identifiez maintenant toutes les espèces différentes présentes à Grenoble.
# N’oubliez pas que les espèces sont une sous catégorie de genre botanique !
# (l’espèce est ‘Platanus Acerifolia’ et pas uniquement ‘Acerifolia’)
data_ESPECE=[]
data_GENRE_ESPECE=[]
ind=data[0].index('ESPECE')
for e in data[1:]:
    if e[ind]!="":
        data_ESPECE.append(e[ind])
        if e[ind-1]!="":
            data_GENRE_ESPECE.append(e[ind-1]+" "+e[ind])
set_ESPECE=set(data_ESPECE)
set_GENRE_ESPECE=set(data_GENRE_ESPECE)
print(len(set_GENRE_ESPECE)) #sans compter le genre vide "" et espece vide ""
```

```python
# En utilisant la structure de données Dict compter le nombre de représentants de
# chaque genre botanique : clé=genre, valeur=nbrT
dict_GENRE_NBR={}
for genre in set_GENRE_BOTA:
    dict_GENRE_NBR[genre]=data_GENRE_BOTA.count(genre)
dict_GENRE_NBR
```

```python
# ○ Quel est le genre botanique le plus représenté ?
def dict_inverse(dic):
    dic_inverse={}
    lGenre=[]
    for nbr in range(1,5277): #à changer par max du dic ?
        lGenre=[]
        for genre in dic:
            if dic[genre]==nbr:
                lGenre.append(genre)
        if lGenre!=[]:
            dic_inverse[nbr]=lGenre
    return dic_inverse

dict_NBR_GENRE=dict_inverse(dict_GENRE_NBR)
dict_NBR_GENRE
```

```python
# ○ Listez tous les représentants uniques de leurs genre botanique
print(dict_NBR_GENRE[1])
```

```python
# ○ Classer tous les genres botaniques en fonction de leur nombre (ordre
# décroissant)
list_GENRE_NBR=list(dict_GENRE_NBR.items())
list_GENRE_NBR.sort(key=lambda a: a[1],reverse=True)
print(list_GENRE_NBR)
```

```python
# ○ Quelle est la proportion du genre botanique le plus représenté parmi tous les
# genres ?
total=0
for e in list_GENRE_NBR:
    total+=e[1]
print(list_GENRE_NBR[0][1]/total*100)
```

```python
# ○ Quelle est la proportion de chacun des genres botaniques par rapport au total ?
l_prop_GENRE=[]
for i in range(len(list_GENRE_NBR)):
    print(list_GENRE_NBR[i][1]/total*100)
    l_prop_GENRE.append(list_GENRE_NBR[i][1]/total*100)
```
